```python
import numpy as np
import matplotlib.pyplot as plt

# create sin wave of period = 2
t = np.linspace(0, 10, 1000)
x = (np.sin(t * np.pi) * 100 + 100).astype(np.uint)

# plot
plt.figure(figsize=(10, 5))
plt.plot(t, x, label='Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# save data
f = open('sin_wave.dat', 'w')
for i in range(len(x)):
    print(x[i], file=f)
f.close()
```

```python
def autocorrelation(signal):
    length = len(signal)
    autocorr = []

    for delay in range(length):
        signal_original = signal[delay:]
        signal_delayed = signal[:length - delay]
        auto = np.sum(signal_original * signal_delayed).astype(np.uint) // length
        autocorr.append(auto)
    
    return autocorr

# compute ACF
acf = autocorrelation(x)

# plot
plt.figure(figsize=(10, 5))
plt.plot(t, acf, label='ACF')
plt.xlabel('Lagged Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# save data
f = open('sin_wave_golden.dat', 'w')
for i in range(len(acf)):
    print(acf[i], file=f)
f.close()
```
